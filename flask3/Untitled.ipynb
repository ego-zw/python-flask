```python
import pandas as pd
import numpy as np
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType,ThemeType
from pyecharts.charts import Bar,Tab,Line,Map,Timeline
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Funnel, Page
```

```python
df = pd.read_csv('gdp.csv',encoding='utf8',index_col=0)
df1 = pd.read_csv('pgdi.csv',encoding='utf8',index_col=0)
df2 = pd.read_csv('cpi.csv',encoding='utf8',index_col=0)
df3 = pd.read_csv('rengdp.csv',encoding='utf8',index_col=0)
```

### （1）分省GDP水平

```python
df
```

```python
list(df.index)
```

```python
for i in range(2013, 2018):
    print(list(zip(list(df.index),list(df["{}年".format(i)]))))
```

```python

def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2013, 2018):
        map0 = (
            Map()
            .add(
                "分省GDP/亿元",list(zip(list(df.index),list(df["{}年".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}dependency ratio".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_= 800,max_=90000),
            
            )
        )
        tl.add(map0,"{}年".format(i))
    return tl
timeline_map().render_notebook()
```

### 数据分析
分省年度GDP数据展现了各省GDP数值差距之大，GDP数值最小的西藏从2013年815亿元增长到2017年的1310亿元，数据显示了两个现象，1.西藏自治区的GDP在全国处于低水平。2.西藏自治区GDP在不断的增长中。GDP数值最大的广东省从2013年62474亿元到2017年增长到89705亿元，广东省的GDP也在在不断的增长。
### 数据得出结论：
中国分省GDP差距大，同时中国各省GDP在不断的增长，中国的经济不断向前发展着。

### （2）分省人均GDP水平

```python
df3
```

```python
for i in range(2013, 2018):
    print(list(zip(list(df1.index),list(df1["{}年".format(i)]))))
```

```python
def funnel_base() -> Funnel:
    tl = Timeline()
    for i in range(2013,2018):
        c = (
        Funnel()
        .add("分省人均GDP/元", list(zip(list(df3.index),list(df3["{}年".format(i)]))))
        .set_global_opts(
                title_opts=opts.TitleOpts(title="{}分省人均GDP/元".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_= 23000,max_=130000)
            
            )
        )
        tl.add(c,"{}年".format(i))
    return tl
```

```python
funnel_base().render_notebook()
```

### 现象分析
+ 历年数据，中国分省人均GDP不断增长。以2017年为例可以看出中国分省人均GDP水平分布不均，最高的上海市可以达到126634元，最低的甘肃仅有28497元。

### 结论：
+ 中国经济的发展，在人均GDP创造上可以体现，同时在经济高速发展的过程中，也造成了不平衡的状态，以甘肃、西藏、云南和贵州等偏远地区享受到的发展速度远远低于东部沿海地区，中国经济发展展现了不平衡的状态。
+ 地区发展的不平衡始终是中国经济发展的缺陷，缩小地区发展差距，提高全民幸福感应该是国家的要务。

### （3）分省居民人均可支配收入

```python
df1
```

```python

for i in range(2013, 2018):
    print(list(zip(list(df2.index),list(df2["{}年".format(i)]))))
```

```python
def funnel_base() -> Funnel:
    tl = Timeline()
    for i in range(2013,2018):
        c = (
        Funnel()
        .add("人均可支配收入/元", list(zip(list(df1.index),list(df1["{}年".format(i)]))))
        .set_global_opts(
                title_opts=opts.TitleOpts(title="{}人均可支配收入/元".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_= 9000,max_=65000)
            
            )
        )
        tl.add(c,"{}年".format(i))
    return tl
```

```python
funnel_base().render()
```

### 现象分析 
+ 各省人均可支配收入都在不断地增长。
+ 分省人均可支配收入差距巨大，最多的地区和最少的地区差不多有四倍多的差距。

### 结论
+ 从中国GDP和分省人均GDP数据 中，我们已经得出了中国经济向前发展的结论。在人均可支配收入数据中，分省人均可支配收入的增长显示了中国的经济发展落实到了人民群众的身上，人民群众享受到了中国经济发展带来的成就。
+ 危机的存在：中国经济在不断的发展，人民群众的收入也在增长是不可否认的事实，但是不容忽视的是人均的差距，在分省人均可支配收入中，足足四倍的差距仿佛一根刺，扎在了中国经济发展的动脉上。中国要实现两个“一百年目标”要保持经济增长的同时，也不可忽视人均差距的存在。

+ 各省居民消费水平差距巨大，最大差距达5至6倍之大。

### 结论
+ 居民消费水平真正体现了居民享受经济发展的成果水平，结合人均GDP创造数据，可以看出中国经济发展的创造和经济发展成果的享受呈现了同比增长的趋势。
+ 原本经济发展较落后的地区在持续的经济增长态势中，依旧处于落后的趋势，如果不加以干预，落后地区与发达地区的差距会愈加扩大，容易造成社会分层，带来社会不安因素。
